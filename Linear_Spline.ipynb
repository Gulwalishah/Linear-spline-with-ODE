```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
t_data = np.array([-2, -1, 0, 1, 2 ,6])
y_data = np.array([100, 36.79, 13.49, 4.94, 1.81,2.7])

# Time at which we want to estimate y
t_to_estimate = 2.5

# Linear interpolation
y0 = y_data[0]
k = -np.log(y_data[1] / y0)
estimated_y = y0 * np.exp(-k * t_to_estimate)

# Create a plot
plt.plot(t_data, y_data, 'bo-', label='Data Points')
plt.plot(t_to_estimate, estimated_y, 'ro', label='Interpolated Point')
plt.xlabel('Time (t)')
plt.ylabel('Quantity (y)')
plt.title('Linear Interpolation for Differential Equation')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Display the estimated quantity
print(f"Estimated quantity at t = {t_to_estimate}: {estimated_y}")
```
